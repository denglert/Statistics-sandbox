# Correlated systematics

## Uncorrected, raw measurements:

\begin{equation}
\begin{aligned}
    x_{1}  &\sim Poi(a_{1} + \beta_{1} \nu) = \mathcal{L}_{1} \\
    x_{2}  &\sim Poi(a_{2} + \beta_{2} \nu) = \mathcal{L}_{2} \\
    \alpha &\sim \mathcal{N} (\nu, \sigma) = \mathcal{L}_{\alpha}
\end{aligned}
\end{equation}

\begin{equation}
    \mathbf{\mathcal{L}}_{\text{comb}}
    =
    \mathcal{L}_{1} \mathcal{L}_{2} \mathcal{L}_{\alpha}
\end{equation}

## Corrected measurements

\begin{equation}
\begin{aligned}
    y_{1} &= x_{1} - \beta_{1} \alpha \\
    y_{2} &= x_{2} - \beta_{2} \alpha
\end{aligned}
\end{equation}

```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from matplotlib_utils.style import load_mpl_style
```

```python
load_mpl_style('2_by_2.mplstyle')
```

```python
nsamples = 10000
```

```python
a1, a2 = 10, 20
b1, b2 = 3, 3
```

```python
nu, sigma = 5.0, 2.0
```

```python
l1 = a1 + b1 * nu
l2 = a2 + b2 * nu
```

```python
x1s = scipy.stats.poisson.rvs(mu=l1, size=nsamples)
x2s = scipy.stats.poisson.rvs(mu=l2, size=nsamples)
alphas = scipy.stats.norm.rvs(loc=nu, scale=sigma, size=nsamples)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax = ax.flatten()
ax[0].scatter(x1s, x2s)
ax[0].set_xlabel(r"$x_{1}$")
ax[0].set_ylabel(r"$x_{2}$")
ax[1].scatter(x1s, alphas)
ax[1].set_xlabel(r"$x_{1}$")
ax[1].set_ylabel(r"$\alpha$")
ax[2].scatter(x2s, alphas)
ax[2].set_xlabel(r"$x_{2}$")
ax[2].set_ylabel(r"$\alpha$")
fig.delaxes(ax[3])
```

```python
y1s = x1s - b1*alphas
y2s = x2s - b2*alphas
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax = ax.flatten()
ax[0].scatter(y1s, y2s)
ax[0].set_xlabel(r"$y_{1}$")
ax[0].set_ylabel(r"$y_{2}$")
ax[1].scatter(y1s, alphas)
ax[1].set_xlabel(r"$y_{1}$")
ax[1].set_ylabel(r"$\alpha$")
ax[2].scatter(y2s, alphas)
ax[2].set_xlabel(r"$y_{2}$")
ax[2].set_ylabel(r"$\alpha$")
fig.delaxes(ax[3])
```
