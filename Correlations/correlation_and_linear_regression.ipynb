```python
import scipy.stats
```

```python
mu1 = 0.0
mu2 = 0.0
sigma1 = 1.0
sigma2 = 0.99
rho = 0.6
```

```python
sigma12 = sigma1*sigma2*rho
```

```python
mu = np.array([mu1, mu2])
cov = np.array([[sigma1**1, sigma12**2], [sigma12**2, sigma2]])
```

Angle of the major axis $\phi$ is given through the relation

\begin{equation}
    \tan 2 \phi
    = 
    \frac{2 \rho \sigma_{1} \sigma_{2}}{\sigma^{2}_{1} - \sigma^{2}_{2}}
\end{equation}

```python
tan2phi = 2.0*rho*sigma1*sigma2/(sigma1**1 - sigma2**2)
```

```python
phi = np.arctan(tan2phi)/2.0
```

```python
tanphi = np.tan(phi)
tanphi
```

```python
X = scipy.stats.multivariate_normal.rvs(mean=mu, cov=cov, size=5000)
```

```python
f,a = plt.subplots()
a.scatter(X[:,0], X[:,1], alpha=0.1)
```

```python
x = X[:,0]
y = X[:,1]
```

```python
np.cov(m=(x,y))
```

```python
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x,y)
```

```python
print("Slope: ", slope)
print("intercept: ", intercept)
print("r-value: ", r_value)
print("p-value: ", p_value)
print("std_err: ", std_err)
```

```python
x_lr = np.linspace(-5.0, 5.0, 100)
y_lr = intercept + slope*x_lr
```

```python
f,a = plt.subplots()
a.scatter(X[:,0], X[:,1], alpha=0.1)
a.plot(x_lr, y_lr, c='k')
```
