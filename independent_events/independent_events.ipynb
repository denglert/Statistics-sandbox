# Independent events - factorisability

\begin{equation}
    p()
\end{equation}

```python
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nsmall = 5
nmedium = 20
nlarge = 50
```

```python
small = scipy.stats.poisson.rvs(mu=10,  size=nsmall)
medium = scipy.stats.poisson.rvs(mu=10, size=nmedium)
large = scipy.stats.poisson.rvs(mu=10,  size=nlarge)
```

```python
lambdas = np.linspace(5, 20, 100)
```

```python
logpdf_small = np.array([np.sum(-scipy.stats.poisson.logpmf(small, mu=l)) for l in lambdas])
logpdf_medium = np.array([np.sum(-scipy.stats.poisson.logpmf(medium, mu=l)) for l in lambdas])
logpdf_large = np.array([np.sum(-scipy.stats.poisson.logpmf(large, mu=l)) for l in lambdas])
```

```python
fig, ax = plt.subplots()
ax.plot(lambdas, logpdf_small, label='$n =$ {}'.format(nsmall))
ax.plot(lambdas, logpdf_medium, label='$n =$ {}'.format(nmedium))
ax.plot(lambdas, logpdf_large, label='$n=$ {}'.format(nlarge))
ax.set_xlabel(r'$\lambda$')
ax.set_ylabel(r'$- \ln \mathcal{L} - (- \ln \mathcal{L}_{max})$')
#ax.set_ylim(0.0, 50.0)
ax.legend();
```

```python
fig, ax = plt.subplots()
ax.plot(lambdas, logpdf_small-logpdf_small.min(), label='$n =$ {}'.format(nsmall))
ax.plot(lambdas, logpdf_medium-logpdf_medium.min(), label='$n =$ {}'.format(nmedium))
ax.plot(lambdas, logpdf_large-logpdf_large.min(), label='$n=$ {}'.format(nlarge))
ax.set_xlabel(r'$\lambda$')
ax.set_ylabel(r'$- \ln \mathcal{L} - (- \ln \mathcal{L}_{max})$')
ax.set_ylim(0.0, 50.0)
ax.legend();
```

## Combination

```python
lambdas = np.linspace(1, 20, 100)
```

```python
observations = scipy.stats.poisson.rvs(mu=10,  size=3)
```

```python
logpdf_combined = np.array([np.sum(-scipy.stats.poisson.logpmf(observations, mu=l)) for l in lambdas])
```

```python
fig, ax = plt.subplots()
for n in observations:
    logpdf = np.array([-scipy.stats.poisson.logpmf(n, mu=l) for l in lambdas])
    label = "n = {}".format(n)
    ax.plot(lambdas, logpdf, label=label)
ax.plot(lambdas, logpdf_combined, linewidth=3)
ax.legend()
```

```python
fig, ax = plt.subplots()
for n in observations:
    logpdf = np.array([-scipy.stats.poisson.logpmf(n, mu=l) for l in lambdas])
    label = "n = {}".format(n)
    ax.plot(lambdas, logpdf-logpdf.min(), label=label)
ax.plot(lambdas, logpdf_combined-logpdf_combined.min(), linewidth=3)
ax.set_xlim(1.0, 20.0)
ax.set_ylim(0.0, 30.0)
ax.legend()
```
