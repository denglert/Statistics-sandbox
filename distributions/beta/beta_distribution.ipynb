# Beta distribution

**References:**
- https://en.wikipedia.org/wiki/Beta_distribution

In probability theory and statistics, the beta distribution is a family of continuous probability distributions

Defined on the interval:
- [0, 1]

Parametrized by two positive shape parameters, denoted by:
- $α$
- $β$

that appear as exponents of the random variable and control the shape of the distribution.

The beta distribution has been applied to model the behavior of random variables limited to intervals of finite length in a wide variety of disciplines.

**In Bayesian inference:**

The beta distribution is the conjugate prior probability distribution for the:
- Bernoulli
- binomial
- negative binomial
- geometric distributions.

For example, the beta distribution can be used in Bayesian analysis to describe initial knowledge concerning probability of success such as the probability that a space vehicle will successfully complete a specified mission. The beta distribution is a suitable model for the random behavior of percentages and proportions.

The usual formulation of the beta distribution is also known as the beta distribution of the first kind, whereas beta distribution of the second kind is an alternative name for the beta prime distribution. 

## Probability distribution function (PDF)

\begin{equation}
    f(x)
    =
    \frac{x^{\alpha-1} (1-x)^{\beta-1}}{B(\alpha,\beta)}
\end{equation}

where $B(\alpha,\beta)$ is the beta function:

\begin{equation}
    B(\alpha,\beta)
    =
    \frac{\Gamma(\alpha) \Gamma(\beta)}{\Gamma(\alpha + \beta)}
\end{equation}

\begin{equation}
    B(\alpha, \beta)
    =
    \int_{0}^{1}
    dt
    ~
    t^{\alpha-1}
    (1-t)^{\beta-1}
\end{equation}

## Moments

### Mean - $E[X]$

\begin{equation}
    E[X] = \frac{\alpha}{\alpha+\beta}
\end{equation}

### Variance - $Var[X]$

\begin{equation}
    V[X] = \frac{\alpha \beta}{ (\alpha+\beta)^{2} (\alpha + \beta +1)}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cm as cmx
import scipy.stats
%matplotlib inline
```

```python
x = np.linspace(0.0, 1.0, 100)
```

```python
pdf_alpha_beta = {}
```

```python
alphas = np.linspace(0.0, 2.0, 5)
betas = np.linspace(0.0, 2.0, 5)
```

```python
alphas
```

```python
for alpha in alphas:
    pdf_alpha_beta[alpha] = {}
    for beta in betas:
        pdf_alpha_beta[alpha][beta] = scipy.stats.beta.pdf(x, a=alpha, b=beta)
```

```python
blues = cm = plt.get_cmap('Blues') 
blues_color_norm  = colors.Normalize(vmin=min(alphas), vmax=max(alphas))
blues_color_map   = cmx.ScalarMappable(norm=blues_color_norm, cmap=blues)
```

```python
reds = cm = plt.get_cmap('Reds') 
reds_color_norm  = colors.Normalize(vmin=min(betas), vmax=max(betas))
reds_color_map   = cmx.ScalarMappable(norm=reds_color_norm, cmap=reds)
```

## PDF

```python
f,axes = plt.subplots(figsize=(10,6), ncols=2)
a = axes.flatten()
beta = 1.0
for alpha in alphas:
    label = r"$\alpha = {:.1f}$, $\beta$= {:.1f}".format(alpha, beta)
    c = blues_color_map.to_rgba(alpha)
    a[0].plot(x,pdf_alpha_beta[alpha][beta], label=label, c=c)
a[0].legend()

alpha = 1.0
for beta in betas:
    label = r"$\alpha = {:.1f}$, $\beta$= {:.1f}".format(alpha, beta)
    c = reds_color_map.to_rgba(beta)
    a[1].plot(x,pdf_alpha_beta[alpha][beta], label=label, c=c)
a[1].legend()
```

```python
f,a = plt.subplots(figsize=(8,6))
alpha = 0.5
beta = 0.5
label = r"$\alpha = {:.1f}$, $\beta$= {:.1f}".format(alpha, beta)
a.plot(x, pdf_alpha_beta[alpha][beta], label=label)
alpha = 2.0
beta = 2.0
label = r"$\alpha = {:.1f}$, $\beta$= {:.1f}".format(alpha, beta)
a.plot(x, pdf_alpha_beta[alpha][beta], label=label)
a.legend()
```
