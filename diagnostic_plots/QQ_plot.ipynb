# Q-Q plot

**References:**
- https://en.wikipedia.org/wiki/Q%E2%80%93Q_plot

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
npts = 100
```

```python
x_norm  = np.random.normal(loc = 20, scale = 5, size=npts)   
x_binom = np.random.binomial(n=50,  p=20/50,    size=npts)
x_exp  = np.random.exponential(scale=20,    size=npts)
```

```python
f,ax = plt.subplots()
ax.hist(x_binom)
```

```python
f,ax = plt.subplots()
ax.hist(x_exp)
```

```python
f,ax = plt.subplots()
(osm, osr), (slope, intercept, r) = scipy.stats.probplot(x_norm,  dist="norm", plot=ax)
```

```python
f,ax = plt.subplots()
(osm, osr), (slope, intercept, r) = scipy.stats.probplot(x_binom, dist="norm", plot=ax)
```

```python
f,ax = plt.subplots()
(osm, osr), (slope, intercept, r) = scipy.stats.probplot(x_exp, dist="norm", plot=ax)
```
